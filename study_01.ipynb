### 함수 기본
* 전역 변수는 함수 내에서만 사용할 수 있음(메모리 상에서 제거)

```python
def my_sum(a):
    sums = 0
    for i in a:
        sums += i
    return sums
a = [1,2,3,4,5]
print(my_sum(a))
```

```python
def my_sum(*a):
    sums = 0
    for i in a:
        sums += i
    return sums
user1 = 1
user2 = 2
user3 = 3
print(my_sum(user1,user2,user3))
```

### **키워드, 매개변수 kwargs**

* 인공지능 함수에 x, y data의 개수가 정해지지 않았기 때문에 **kwargs 많이 사용 
* {x : y}의 딕셔너리 구조로 입력값으로 들어옴 - 갯수의 쌍이 일치해야 함

```python
def print_kwargs(**kwargs):
    print(kwargs)
print_kwargs(name="foo", age="3")
# 복수의 인자값이 (** + keyword arguments)에서 하나의 값으로 입력된다. 
```

```python
def introduceEnglishName(**kwargs):
    for key,value in kwargs.items():
        print(kwargs)
        if 'ant' in kwargs.keys():
            print('주인님 오셨군요, 오늘 기분이 어떠세요?')
        else:
            print("{0} is {1}".format(key,value))

introduceEnglishName(myName="chris")
introduceEnglishName(ant="chris")    
```

### **딕셔너리(kwargs)를 사용하는 이유**
* x와 y데이터의 개수가 다르기 때문에
* 딕셔너리 구조에서 개수를 맞추어 입력값으로 사용하기 위함
    * { xData_input data : yData_label }  

```python
label = ['dog', 'cat']

img = [
[[[255,0,0],[255,0,0],[255,0,0],[255,0,0]], # pixel 1,2,3,4
[[255,0,0],[5,0,0],[0,0,0],[2,50,0]]],

[[[255,0,0],[255,0,0],[0,0,255],[255,0,0]],
[[255,0,0],[50,0,250],[255,0,0],[255,0,0]]]
]

def DicImgListShow(user = 'bu', **kwargs):
    xData = kwargs['xData']
    yData = kwargs['yData']
    
    for x, y in zip(xData, yData):
        plt.imshow(x)
        plt.title(y)
        plt.show()

DicImgListShow(xData = img, yData = label)
```
